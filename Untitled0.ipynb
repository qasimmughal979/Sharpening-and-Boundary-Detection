```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Let user upload an image
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Read the uploaded image in grayscale
img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

# Laplacian Filter (Second Derivative)
laplacian = cv2.Laplacian(img, cv2.CV_64F, ksize=3)
laplacian_8u = cv2.convertScaleAbs(laplacian)

# Original - Laplacian
original_minus_laplacian = cv2.subtract(img, laplacian_8u)

# Display Results for Laplacian
titles_laplacian = [
    "Original",
    "Laplacian (2nd derivative)",
    "Original - Laplacian"
]
images_laplacian = [img, laplacian_8u, original_minus_laplacian]

plt.figure(figsize=(15, 5))
for i in range(len(images_laplacian)):
    plt.subplot(1, 3, i+1)
    plt.imshow(images_laplacian[i], cmap="gray")
    plt.title(titles_laplacian[i])
    plt.axis("off")
plt.show()

# Sobel Filters (First Derivative)
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
sobel_combined_8u = cv2.convertScaleAbs(sobel_combined)

# Original - Sobel Combined
original_minus_sobel = cv2.subtract(img, sobel_combined_8u)


# Display Results for Sobel
titles_sobel = [
    "Original",
    "Sobel X (1st derivative)",
    "Sobel Y (1st derivative)",
    "Sobel Combined",
    "Original - Sobel Combined"
]
images_sobel = [img, cv2.convertScaleAbs(sobel_x), cv2.convertScaleAbs(sobel_y), sobel_combined_8u, original_minus_sobel]

plt.figure(figsize=(20, 5))
for i in range(len(images_sobel)):
    plt.subplot(1, 5, i+1)
    plt.imshow(images_sobel[i], cmap="gray")
    plt.title(titles_sobel[i])
    plt.axis("off")
plt.show()


# ---- Kernels for Report ----
# Sobel X Kernel:
# [[-1, 0, 1],
#  [-2, 0, 2],
#  [-1, 0, 1]]

# Sobel Y Kernel:
# [[-1, -2, -1],
#  [ 0,  0,  0],
#  [ 1,  2,  1]]

# Laplacian Kernel (3x3):
# [[ 0,  1, 0],
#  [ 1, -4, 1],
#  [ 0,  1, 0]]
```

## Implement smoothing filters

### Subtask:
Add code to apply Mean, Median, and Mode filters to the uploaded image.


**Reasoning**:
Apply Mean, Median, and Mode filters to the image as instructed.



```python
# Apply Mean filter (5x5 kernel)
mean_smoothed = cv2.blur(img, (5, 5))

# Apply Median filter (5x5 kernel)
median_smoothed = cv2.medianBlur(img, 5)

# Apply Mode filter (5x5 kernel) - Manual Implementation
# Create an empty image to store the mode smoothed result
mode_smoothed = np.zeros_like(img)

# Get image dimensions
rows, cols = img.shape

# Define kernel size
ksize = 5
pad = ksize // 2

# Pad the image to handle borders
img_padded = cv2.copyMakeBorder(img, pad, pad, pad, pad, cv2.BORDER_REPLICATE)

# Iterate over each pixel in the original image
for i in range(rows):
    for j in range(cols):
        # Extract the neighborhood
        neighborhood = img_padded[i:i+ksize, j:j+ksize]
        # Calculate the mode of the neighborhood
        values, counts = np.unique(neighborhood, return_counts=True)
        mode_smoothed[i, j] = values[np.argmax(counts)]

# Display the results
titles_smoothing = [
    "Original",
    "Mean Smoothed",
    "Median Smoothed",
    "Mode Smoothed"
]
images_smoothing = [img, mean_smoothed, median_smoothed, mode_smoothed]

plt.figure(figsize=(20, 5))
for i in range(len(images_smoothing)):
    plt.subplot(1, 4, i+1)
    plt.imshow(images_smoothing[i], cmap="gray")
    plt.title(titles_smoothing[i])
    plt.axis("off")
plt.show()
```

## Introduce noise

### Subtask:
Generate synthetic images with different types of noise (e.g., Gaussian, salt-and-pepper) to test the filters.


**Reasoning**:
Implement a function to add Gaussian and salt-and-pepper noise to an image, then apply it to the original image and display the results.



```python
def add_noise(image, noise_type="gaussian", mean=0, std_dev=25, sp_amount=0.05):
    """Adds noise to an image."""
    if noise_type == "gaussian":
        gaussian_noise = np.random.normal(mean, std_dev, image.shape).astype('uint8')
        noisy_image = cv2.add(image, gaussian_noise)
        noisy_image = np.clip(noisy_image, 0, 255)
    elif noise_type == "salt-and-pepper":
        noisy_image = np.copy(image)
        total_pixels = image.size
        # Add salt noise (white pixels)
        num_salt = np.ceil(sp_amount * total_pixels * 0.5)
        coords_salt = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
        noisy_image[coords_salt[0], coords_salt[1]] = 255
        # Add pepper noise (black pixels)
        num_pepper = np.ceil(sp_amount * total_pixels * 0.5)
        coords_pepper = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
        noisy_image[coords_pepper[0], coords_pepper[1]] = 0
    else:
        noisy_image = np.copy(image)  # Return original image if noise type is unknown
    return noisy_image

# Add Gaussian noise to the original image
img_gaussian_noise = add_noise(img, noise_type="gaussian", std_dev=25)

# Add Salt-and-Pepper noise to the original image
img_sp_noise = add_noise(img, noise_type="salt-and-pepper", sp_amount=0.05)

# Display the original and noisy images
titles_noise = [
    "Original",
    "Image with Gaussian Noise",
    "Image with Salt-and-Pepper Noise"
]
images_noise = [img, img_gaussian_noise, img_sp_noise]

plt.figure(figsize=(15, 5))
for i in range(len(images_noise)):
    plt.subplot(1, 3, i+1)
    plt.imshow(images_noise[i], cmap="gray")
    plt.title(titles_noise[i])
    plt.axis("off")
plt.show()
```
